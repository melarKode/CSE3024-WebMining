```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
sample = str(input('Enter Sample Text- '))
```

```python
#Removing Stop words
stopwords = set(stopwords.words('english'))
wordTokens = word_tokenize(sample)

filteredSample = [w for w in wordTokens if w not in stopwords]
print(len(wordTokens))
print(len(filteredSample))
```

```python
#Removing punctuations
noPunctuations = [w for w in filteredSample if w.isalpha()]
print(len(noPunctuations))
```

```python
preStemming = [w.lower() for w in noPunctuations]
```

```python
#Stemming
from nltk import PorterStemmer

porter = PorterStemmer()
answer = [porter.stem(w) for w in preStemming]
```

```python
print(answer)
```

```python
freq1 = nltk.FreqDist(wordTokens)
freq2 = nltk.FreqDist(filteredSample)
freq3 = nltk.FreqDist(noPunctuations)
freq4 = nltk.FreqDist(preStemming)
freq5 = nltk.FreqDist(answer)
```

```python
freq1.plot(25,cumulative=False)
freq2.plot(25,cumulative=False)
freq3.plot(25,cumulative=False)
freq4.plot(25,cumulative=False)
freq5.plot(25,cumulative=False)
```
